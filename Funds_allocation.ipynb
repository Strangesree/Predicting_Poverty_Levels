```python
import pandas as pd

# Create a DataFrame for the sub-regions in Tumkur
data1 = pd.DataFrame({
    'Sub-Region': ['Gubbi', 'Tumkur Rural', 'Tiptur', 'Pavgada', 'Chikkanayakanahalli', 'Turvekere', 'Sira', 'Tumkur Urban'],
    'Education': [0.75, 0.65, 0.55, 0.45, 0.85, 0.75, 0.65, 0.30],
    'Food': [0.80, 0.70, 0.60, 0.50, 0.90, 0.80, 0.70, 0.45],
    'Sanitation': [0.70, 0.60, 0.50, 0.40, 0.80, 0.70, 0.60, 0.35],
    'Healthcare': [0.85, 0.75, 0.65, 0.55, 0.95, 0.85, 0.75, 0.20]
})
data1.to_csv('data.csv', index=False)
```

```python
data = pd.read_csv('/content/data.csv')
data['Poverty Index'] = data[['Education', 'Food', 'Sanitation', 'Healthcare']].mean(axis=1)
```

```python
data_sorted = data.sort_values(by='Poverty Index', ascending=False)
data['Rank'] = range(1, len(data_sorted) + 1)
data
```

```python
#Funds Distribution
# Create a DataFrame with the poverty indexes of the sub-regions
data2 = pd.DataFrame({
    'Sub-Region': ['Gubbi', 'Tumkur Rural', 'Tiptur', 'Pavgada', 'Chikkanayakanahalli', 'Turvekere', 'Sira', 'Tumkur Urban'],
    'Poverty Index': [0.75, 0.65, 0.55, 0.45, 0.85, 0.75, 0.65, 0.30]
})

data2.to_csv('data2.csv',index = False)
```

```python
data2 = pd.read_csv('/content/data2.csv')
# Calculate the allocation amount for each sub-region based on the poverty index (in lakhs)
data2['Allocation'] = data2['Poverty Index'] / data2['Poverty Index'].sum() * 10000000 / 100000
```

```python
data2
```

```python
# Update the poverty index based on the improvement after funds allocation
data2['Improved Poverty Index'] = data2['Poverty Index'] - data2['Allocation'] * 0.01
data2
```

```python
data2.columns = ['Sub-Region','Poverty Index','Before Allocation', 'After Allocation']
data2
```
